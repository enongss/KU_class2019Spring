```python
model = Sequential()
model.add(Dense(10, input_shape=[13]))
model.add(Dense(20))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
# 위에서 Hidden layer를 넣은 것
# model.add(Dense(10, input_shape=[13])) : 13개의 입력부분, 10개의 Hidden layer부분
# model.add(Dense(20)) : 중간에 있는 20개의 Hidden layer부분
# model.add(Dense(1)) : 1개의 출력부분 (마지막은 무조건 1!!!)
# 질문 : 여기서 Hidden layer의 개수는? 2개!!! 1개 아님!!!
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
# epochs = 숫자 : 데이터를 사용하는 숫자 (한번 쓸 때 404개를 다 씀)
# Hidden layer 있는 것과 없는 것 비교 => 별로 차이가 없음. 왜? linear한 것은 별로 차이가 없기 때문! => nonlinearity를 부여해야 함!
```

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation = 'relu'))
model.add(Dense(20))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
# nonlinearity를 추가하기 위해 처음 Hidden layer에 activation = 'relu'를 추가함!

# 여기서 중요한 점! 입력값이 13개, 출력값이 1개, Hidden layer는 2개
# 출력부분에서 140, 220, 21이 어떻게 나왔는지 반드시 알아야 함!
# 140은 입력 부분 13개 + 1로 된 node와 그 다음 Hidden layer에 있는 node의 개수를 곱한 것
# 220은 이전 Hidden layer의 node(***여기서 1이 있지만 이전에는 1에 화살표 연결하지 않음!)와 다음 Hidden layer의 node 개수를 곱한 것
# 21은 마지막 Hidden layer의 node(***역시 여기서 1을 포함!) 개수와 출력 부분 1개를 곱한 것
# activation(sigmoid, tanh, relu)을 Hidden layer에 부여하면 성능이 올라감. 그렇다면 출력부분은? regression일 때 아무것도 주지 말고, classification일 때는 경우를 나누어야 함
# classification에서는 sigmoid와 softmax가 있는데 sigmoid는 binary, softmax는 multiple choice일때 부여함
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
# nonlinearity를 추가하니까 확실히 loss가 줄어듦!
```

```python
model.evaluate(x_test, y_test)
```

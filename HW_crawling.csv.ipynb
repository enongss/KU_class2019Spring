```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen("https://comic.naver.com/webtoon/weekday.nhn")
html
```

```python
bsObject = BeautifulSoup(html, "html.parser")
bsObject
```

```python
for n in range(0,9):
    li = bsObject.find("li", {"class":"rank0"+str(n+1)})
    title = li.select('a')[0].get_text()
    print(n+1, title)
```

```python
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title'])
    
    for n in range(0,9):
        li = bsObject.find("li", {"class":"rank0"+str(n+1)})
        title = li.select('a')[0].get_text()
        writer.writerow([n+1, title])
```
